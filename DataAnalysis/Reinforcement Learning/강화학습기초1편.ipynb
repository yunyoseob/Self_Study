6. 이 과정을 반복하면서 쥐는 지렛대를 누르면 먹이를 먹을 수 있다는 것을 학습하게 된다. (학습)

# MDP(Markov Decision Process)

위에서 보앗듯, 에이전트가 행동을 해야 환경은 보상을 줍니다. 실사례로 노동자가 일을 해야, 고용자가 월급을 주는 것 처럼 말이죠.

MDP는 이렇게 순차적으로 행동을 결정하는 문제를 정의할 때 사용하는 방법입니다.  MDP는 순차적 행동 결정 문제를 수학적으로 정의해서 에이전트가 순차적 행동 결정 문제에 접근 할 수 있게 합니다.

이렇게 순차적 행동 결정 문제의 구성요소들을 MDP라고 부릅니다.


그렇다면 순차적으로 행동을 결정할 때 그 구성 요소(MDP)는 무엇일까요?

탁구 치는 로봇 영상 사례

탁구를 치는 에이전트 예시: 탁구공의 위치만 알고 속도를 모른다면 에이전트는 사실상 탁구를 칠 수 없음. 에이전트가 탁구 치는 것을 학습하려면 탁구공의 위치, 속도, 가속도 같은 정보가 필요함.


이렇듯 에이전트가 상태를 통해 상황을 판단해서 행동을 결정하기 위해서는  충분한 정보를 제공해야 함. 





2) 행동(A: Action)
--

에이전트가 어떠한 상태에서 취할 수 있는 행동.
에이전트가 행동을 취하면 환경은 에이전트에게 보상을 주고 다음 상태를 알려줌.

Q&A
--

Q. 상태는 어떻게 아나요?

A. 상태는 "자신의 상황에 대한 관찰"이라고 표현 할 수 있는데요. 로봇과 같은 세상에서의 에이전트에게 상태는 센서 값이 될 것입니다. 하지만 보통 에이전트는 사용자가 상태를 정의해줘야 합니다.가령, 알파고 사례로 본다면 상태는 바둑판의 좌표를 의미합니다.

3) 보상(R: Reward)
--

에이전트가 자신이 했던 행동들을 평가할 수 있고, 이로 인해 어떤 행동이 좋은 행동인지 알 수 있음.

Q&A
--

Q. 어떤 행동이 좋은 행동인지 어떻게 아는데요?

A. 행동이 얼마나 좋은지가 행동의 가치가 되고 이것을 큐함수라고 합니다. 큐함수에 대한 내용은 뒤에서 다룰 것입니다.

4) 정책(P: Policy)
--

에이전트가 어떤 행동을 해야 하는지 정해놓은 것, 최적정책은 제일 좋은 정책이며, 에이전트는 최적 정책에 다라 행동했을 때, 보상의 합을 최대로 받을 수 있습니다. 

Q&A
--

Q. 최적정책을 어떻게 구하고 어떻게 아는데요?

A. 가치함수는 에이전트가 어떤 정책이 더 좋은 정책인지 판단하는 기준이며, 정의는 다음과 같습니다. 가치함수는 현재 상태로부터 정책을 따라갔을 때 받을 것이라 예상되는 보상의 합입니다. 한편, 행동이 얼마나 좋은지가 행동의 가치가 되고 이것을 큐함수라고 합니다. 최적의 큐함수와 최적의 가치함수를 통해 최적 정책을 구합니다.

Q. 정책이 구체적으로 뭔데요?

$ P(a_{1}|S_{1}) $ 는 time 1에서 어떤 상태에 있을 때, 어떤 액션을 할지에 대한 확률분포
(그리드 월드 예제로 말하면 위, 아래, 왼쪽, 오른쪽의 확률질량함수(pmf) 를 의미함)

즉, 정책이란 State(상태)에서 어떤 Action(행동)을 할 지에 대한 분포를 의미합니다.

- 순차적으로 행동을 결정하는 문제에 대한 정의: MDP

상태
--

MDP상태.png

- 그리드 월드 예제: 자신의 상황에 대한 관찰

$ S=[{(x1,y1), (x2,y2), (x3,y3), (x4,y4), (x5, y5)}] $

- 상태의 집합

$ St=(1,3) $

- 시간 t에 (1,3)이라는 상태

$ St=s $

- 시간 t에서의 상태

- MDP에서는 상태는 시간에 따라 확률적으로 변함.시간 t에 에이전트가 있을 상태는 전체 상태 중에 하나가 됨.

- 시간 t까지 에이전트가 움직이는 것을 임의 실험이라고 본다면 시간 t에 에이전트가 있을 상태는 확률 변수가 됨.

- ex) 주사위를 던지는 실험은 임의 실험이고 주사위를 던져서 나오는 값이 변수, 이 변수는 확률 값을 가지고 있음. 따라서 이 변수는 확률 변수가 됨.

행동
--

$   At=a $

- 시간 t에서의 행동 a

A={up, down, left, right}  

- 그리드 월드에서 에이전트가 할 수 있는 행동의 집합


행동에 따라 상태가 변함 ex) (3,1) -> (4,1)

보상함수(Reward Function)
--

$ r(s,a) = E[R_{t+1}|S_{t}=s, A_{t}=a] $

- E(Expectation)는 기댓값

- 기댓값은 어떤 정확한 값이 아니라 나오게 될 숫자에 대한 예상임. 상태 s에서 행동 a를 했을 경우에 받을 것이라 예상 되는 숫자

- 기댓값을 사용하는 이유: 보상을 에이전트에게 주는 것은 환경이고 환경에 따라서 같은 상태에서 같은 행동을 취하더라도 다른 보상을 줄 수 도 있기 때문에 이 모든 것을 고려해서 보상함수를 기댓값으로 표현함.

- | 는 조건부 확률 

- t+1은 환경이 에이전트에게 알려주는 시점이므로 일어난 시점+1

- t+1은 에이전트가 환경으로부터 하나의 시간 단위가 지난 다음에 보상을 받기 때문인데, 이 시간 단위를 타임스텝(time step)이라고 함.

상태 변환 확률
--

$ P^a_{ss}=P[S_{t+1}=s'|S_t=s, A_t=a] $

- P: 확률

- 상태 변환 확률은 상태 s에서 행동 a를 취했을 때, 다른 상태 s'에 도달할 확률

할인율
--

간단하게 정리하면, 물가가 올라간다고 할 때, 10년 전의 물가와 지금의 물가는 다름.  다른 예시로, 지금 통장의 돈을 입금한다면, 지금의 돈과 10년 뒤의 돈은 다름. 바로 이자율 때문인데, 이런 이자율을 고려하여 현재 가치로 환산한 것을 할인율이라고 함. 

간단하게 요약하면, 시간에 따라 미래 가치를 현재 가치로 환산한 비율이 할인율임.
(10년 뒤 1만원은 지금 얼마일까?에 대한 환산 비율)

강화학습에서는 보상에 대한 할인율을 적용

$ \gamma \in [0,1] $

- 할인율의 정의

$ \gamma^{k-1}R_{t+k} $

- $ \gamma^{k-1}$: 미래에 받을 보상이 할인되는 가치
- 더 먼 미래에 받은 보상일수록 현재의 에이전트는 더 작은 값으로 받아들임.
- 할인율을 고려한 미래 보상의 현재 가치
- $k$: 현재로부터 지난 시간

Q&A

Q. 할인율이 왜  $ \gamma^{k-1}$이 되나요?

A.

은행이자율이 연 i%라고 할 때, 오늘 1000원을 은행에 넣어두면 1년 후 1100원이 되었다. 그렇다면 연 이자율은 몇 i%(interest_rate)일까?


```python
pv=1000 #현재 가치
fv=1100 #미래 가치

# fv=pv+pv(interest_rate)
# fv=pv(1+interset_rate)

interest_rate=(fv/pv-1)

print(f"이자율은 {round(interest_rate*100, 1)}%입니다.")
```

은행이자율이 연 20%라고 할 때, 오늘 1000원을 은행에 넣어두면 1년 후 1200원이 된다.
그럼 1년 후의 1000원을 오늘의 가치로 계산하면 얼마일까? 

```python
# pv=? #현재 가치
fv=1000 #미래 가치
interest_rate=0.2

# fv=pv*(1+interest_rate)
# pv=fv/(1+interset_rate)

pv=fv/(1+interest_rate)
print(f"이자율이 연 10%라고 할 때, 1년 후의 1000원을 오늘의 가치로 환산하면 {round(pv, 1)}원 입니다.")
```

그렇다면 1년 뒤 미래에 1000원이 현재에 800원이라고 한다면, 할인율은 몇 %일까?

```python
# pv=fv/(1+discount_rate)^1
# 1+discount_rate= fv/pv
pv=800
fv=1000

discount_rate=(fv/pv)-1

print(f"할인율은 {round(discount_rate*100, 1)}%입니다.")
print(f"1년 뒤의 값을 현재 가치로 반환해주기 위한 할인 가치는 {discount_rate}입니다.")
```

이렇듯 할인 가치를 계속해서 시간의 흐름에 따라 제곱의 수를 증가시켜주면서 현재 보상의 값과 미래 보상의 값에 차이를 부여하고

그 값이 $ \gamma$ 라고 한다면,
k라는 시간이 k만큼 지남에 따라 현재와 미래 보상의 차이를 부여($k-1$)하고, 그렇게 되면  보상 앞에 계수가 $ \gamma^{k-1}$ 됨. 

정책의 정의
--

$ \pi(a|s)=P[A_t=a|S_t=s] $

- 시간  t에 $S_t=s$에 에이전트가 있을 때, 가능한 행동중에서 $A_t=a$를 할 확률

- 에이전트가 강화학습을 통해 학습해야 할 것은 수많은 정책 중에서 최적 정책임.

- 최적 정책은 각 상태에서 단 하나의 행동만을 선택함.

- 그러나, 에이전트가 학습하고 있을 때는 정책이 하나의 행동만을 선택하기보다는 확률적으로 여러 개의 행동을 선택할 수 있어야, 다양한 상황에 대해 학습하고 최적 정책을 찾을 수 있을 것.

- 강화학습은 최적 정책을 얻기 위해서 현재의 정책보다 더 좋은 정책을 학습해나가는 것

# 가치함수

앞으로 받을 보상에 대한 개념이 바로 가치함수

- 에이전트 입장에서는 어떤 상황에서 어떤 행동을 하는 것이 좋은지 알 수 없음.

- 따라서 현재 상태에서 앞으로 받을 보상들을 고려해서 선택해야 좋은 선택을 할 수 있음. 앞으로 받을 보상에 대한 개념이 바로 가치 함수


MDP -> 가치함수 -> 행동 선택
--



$ R_{t+1}+R_{t+2}+R_{t+3}+R_{t+4} +R_{t+5}+ ... $

- 일련의 보상들의 단순 합

할인율이 적용되지 않을 경우

$ 0.1+0.1+0.1+..... = \infty $

$ 1+1+1+..... = \infty $

- 시간을 무한대라고 할 때, 단순 합으로는 수치적으로 두 경우를 구분 할 수 없음.

- 쉽게 말하면 기계는 멍청해서 일시불로 상을 받았는지, 할부로 상을 받았는지 모름.

- 지금 받은 보상과 미래에 받을 보상이 똑같이 취급된다면, 시간에 따른 보상을 구분할 수 없음.

- 예시로, 100이라는 보상 1번과 20이라는 보상 5번 받는 것을 구분할 방법이 없음. 

- 사례로 시간에 따라 물가가 계속 상승한다면, 물가가 상승하기 때문에 연봉을 한 번에 받는 것이 월급으로 받는 것보다 이득인데, 할인율을 적용하지 않으면, 에이전트입장에서 이를 알 방법이 없음.

할인율을 적용할 경우

$ R_{t+1}+\gamma*R_{t+2}+\gamma^2*R_{t+3}+\gamma^3*R_{t+4}+\gamma^4*R_{t+5} ...$

- 할인율을 적용한 보상들의 합

$ G_t= R_{t+1}+\gamma*R_{t+2}+\gamma^2*R_{t+3}+\gamma^3*R_{t+4}+\gamma^4*R_{t+5} ... $

- 반환값의 정의

- G는 반환값을 의미하는데, 반환값이란 에이전트가 실제로 환경을 탐험하며 받은 보상의 합

- 쉽게 얘기하면 다같이 놀러가서 돈을 쓸 때, 한 명이 계산한 경우, 이에 대해 받아야 할 정산 값

- 반환값은 MDP로 정의되는 세계에서 에이전트와 환경의 상호작용이 불확실성을 내포하고 있으므로, 특정 상태의 반환값은 에피소드마다 다를 수 있음.

$ G_1= R_{2}+\gamma*R_{3}+\gamma^2*R_{4}+\gamma^3*R_{5}+\gamma^4*R_{6} $

$ G_2 = R_{3}+\gamma*R_{4}+\gamma^2*R_{5}+\gamma^3*R_{6} $

$ G_3= R_{4}+\gamma*R_{5}+\gamma^2*R_{6}$

$ G_4= R_{5}+\gamma*R_{6} $

$ G_5= R_{6} $


- 받을 보상의 정산: 반환 값

- 우항이 점점 주는 이유 $R_{t+1}$의 t가 시간이기 때문에, 시간이 지나면서 우항이 계속 변함

가치 함수
--

$ v(s) = E[G_t|S_t=s] $

- 앞서 보았든, 에이전트는 상황에 따라 특정 상태의 반환값이 다름. 그렇기 때문에 에이전트는 특정 상태의 가치를 반환값에 대한 기댓값으로 특정 상태의 가치를 판단할 수 있음.이것이 바로 가치함수의 개념임.

- 어떠한 상태에 가면 받을 것이라고 예상되는 값. 가치 함수

앞으로 받을 보상에 대한 기댓값인 가치함수

$ v(s)=E[R_{t+1}+ \gamma R_{t+2}+ \gamma ^2 R_{t+3} ...| S_{t}=s] $

$ \gamma R_{t+2}$ 부터 뒤의 항을 $ \gamma $ 로 묶어주고 그것을 반환값의 형태로 표현

$ v(s)=E[R_{t+1}+ \gamma (R_{t+2}+ \gamma R_{t+3}) ...| S_{t}=s] $

$ v(s)=E[R_{t+1}+ \gamma G_{t+1}| S_{t}=s] $

- 그러나 에이전트가 실제로 받는 보상이 아님. 이 보상은 앞으로 받을 것이라 예상하는 보상임.

- 따라서 이 부분을 앞으로 받을 보상에 대한 기댓값인 가치함수로 표현. 반환 값을 가치 함수로 바꿔서 표현하면 아래와 같음.

$ v(s) = E[R_{t+1}+\gamma v(S_{t+1})|S_t=s] $

-  가치 함수로 표현하는 가치함수의 정의

- 가치 함수는 확률변수가 아니라 특정 양을 나타내는 값이므로 소문자로 표현

- 가치 함수를 고려해야 가장 가치가 높은 상태를 선택할 수 있으므로, 중요

- 반환값인 G에서 상태인 S로 바뀌고 타임스텝이 1추가

여기까지는 가치함수를 정의할 때, 정책을 고려하지 않음. 그러나 에이전트가 앞으로 받을 보상에 대해 생각할 때, 정책을 고려하지 않으면 안 됨.

- 상태에서 상태로 넘어갈 때, 에이전트는 무조건 행동을 해야 하고, 각 상태에서 행동을 하는 것이 에이전트의 정책이기 때문

- 따라서, 타임스텝이 추가 될 때마다 지금 에이전트가 정책에 따라 선택할 행동과 상태 변환 확률을 고려해야 하고, 이것을 고려한 방정식을 벨만 기대 방정식(Bellman Expectation Equation)이라고 함.

정책을 고려한 가치함수 (벨만 기대 방정식)
--

$ v_{\pi}(s) = E_{\pi}[R_{t+1}+\gamma v_{\pi}(S_{t+1})|S_t=s] $

- $\pi$ 는 정책

- 벨만 기대 방정식(Bellman Expectation Equation)은 현재 상태의 가치 함수 $ v_{\pi}(s) $와 다음 상태의 가치 함수 
 $ v_{\pi}(S_{t+1}) $ 사이의 관계를 말해주는 방정식

- 강화학습은 벨만 방정식을 어떻게 풀어나가느냐에 대한 스토리

# 큐함수

상태(INPUT) -> 가치함수 -> 받을 보상의 합(OUTPUT)
--

상태 가치함수가 각 상태에 대해 가치를 알려주는 것처럼 각 행동에 대해 가치를 알려주는 함수가 있다면 에이전트는 그 함수의 값만 보고 바로 행동을 선택할 수 있을 것

에이전트는 가치함수를 통해 다음에 어떤 상태로 가야 할지 판단 할 수 있는데, 어떤 상태로 가면 좋을지 판단하고 그 상태로 가기 위한 행동을 따짐.

하지만 다음 상태로 가기 전에 에이전트가 선택한 행동에 따라 즉각적으로 받는 보상이 달라짐. 또한, 왼쪽으로 가는 행동을 했더라도 상태 변환 확률에 따라 왼쪽으로 가지 않을 수 있음. (왼쪽으로 가는 행동을 취해도, 다른 상태로 갈 수도 있음.)

이 때, 각각의 행동에 대해 가치를 알려주는 함수를 행동 가치함수(큐함수)라고 함.

큐함수(Q Function): 어떤 상태에서 어떤 행동이 얼마나 좋은지 알려주는 함수
--






$ q_{\pi}(s,a) $ 

- s는 상태, a는 행동

- 2개의 행동 상태에서 따로 가치함수를 계산

가치 함수와 큐함수 사이의 관계식

$ v_{\pi}(s,a) = \sum_{a \in A} \pi (a|s)q_{\pi} (s,a) $ 

- 각 행동을 했을 때, 앞으로 받을 보상인 큐함수 $ q_{\pi} (s,a) = \pi (a|s) $에 곱합니다.

- 모든 행동에 대해 큐함수와 $ \pi (a|s)$를 곱한 값을 더하면 가치함수가 됩니다.

큐함수의 정의 (큐함수의 벨만 기대 방정식)

$ q_{\pi} (s,a) = E_{\pi}[R_{t+1} + \gamma q{\pi} (S_{t+1}, A_{t+1}) | S_t=s, A_t = a]  $
